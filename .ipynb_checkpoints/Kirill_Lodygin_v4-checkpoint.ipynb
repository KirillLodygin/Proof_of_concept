````python
def analyze_dataset(df, name):
    print(f"\n{name}:")
    print(f"–†–∞–∑–º–µ—Ä: {df.shape}")
    print(f"–ö–æ–ª–æ–Ω–∫–∏: {df.columns.tolist()}")
    print("\n–¢–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö:")
    df.dtypes
    print("\n–ë–∞–∑–æ–≤–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:")
    df.describe(include='all')
    
    print(f"\n–ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è:")
    for col in df.columns:
        missing = df[col].isna().sum()
        if missing > 0:
            print(f"  {col}: {missing} ({missing/len(df)*100:.1f}%)")
    
    return df
````

````python
def identify_dangerous_descriptions(master_df, approval_threshold=0.7):
    children_keywords = [
        'child', 'children', 'kid', 'kids', 'boy', 'boys', 'girl', 'girls',
        'teenager', 'teen', 'baby', 'babies', 'toddler', 'young boy', 'young girl',
        'little boy', 'little girl', 'schoolboy', 'schoolgirl', 'youngster',
        'minor', 'youth', 'adolescent', 'juvenile', 'baby boy', 'baby girl'
    ]
    
    dangerous_descriptions = []
    
    for idx, row in master_df.iterrows():
        description_text = str(row['description_text']).lower()
        
        has_children_keywords = any(keyword in description_text for keyword in children_keywords)
        
        if has_children_keywords:
            approval_rate = row['approval_rate']
            expert_scores = [row['expert1'], row['expert2'], row['expert3']]
            expert_scores = [s for s in expert_scores if not pd.isna(s)]
            
            high_approval = (not pd.isna(approval_rate) and approval_rate >= approval_threshold)
            expert_confirmation = (len(expert_scores) >= 2 and np.mean(expert_scores) >= 2.5)
            
            if high_approval or expert_confirmation:
                dangerous_descriptions.append({
                    'description_id': row['description_id'],
                    'description_text': row['description_text'],
                    'image_file': row['image_file'],
                    'approval_rate': approval_rate,
                    'expert_mean': np.mean(expert_scores) if expert_scores else None,
                    'reason': 'high_approval' if high_approval else 'expert_confirmation'
                })
    
    return pd.DataFrame(dangerous_descriptions)

print("–ê–Ω–∞–ª–∏–∑ –æ–ø–∞—Å–Ω—ã—Ö –æ–ø–∏—Å–∞–Ω–∏–π —Å —Ä–∞–∑–Ω—ã–º–∏ –ø–æ—Ä–æ–≥–∞–º–∏:")
print("-" * 40)

for threshold in [0.5, 0.6, 0.7, 0.8]:
    dangerous_descriptions = identify_dangerous_descriptions(master_df, threshold)
    print(f"–ü–æ—Ä–æ–≥ {threshold}: –Ω–∞–π–¥–µ–Ω–æ {len(dangerous_descriptions)} –æ–ø–∞—Å–Ω—ã—Ö –æ–ø–∏—Å–∞–Ω–∏–π")

dangerous_descriptions = identify_dangerous_descriptions(master_df, approval_threshold=0.7)
print(f"\n–ò—Ç–æ–≥–æ–≤—ã–π —Å–ø–∏—Å–æ–∫ –æ–ø–∞—Å–Ω—ã—Ö –æ–ø–∏—Å–∞–Ω–∏–π: {len(dangerous_descriptions)}")
````

````python
def extract_bad_images_from_descriptions(dangerous_descriptions_df):
    bad_images = set()
    
    for desc_id in dangerous_descriptions_df['description_id']:
        if '#' in str(desc_id):
            image_file = desc_id.split('#')[0]
            bad_images.add(image_file)
    
    return list(bad_images)

bad_images_list = extract_bad_images_from_descriptions(dangerous_descriptions)
print(f"–ù–∞–π–¥–µ–Ω–æ –ø–ª–æ—Ö–∏—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π: {len(bad_images_list)}")
print(f"–ü—Ä–∏–º–µ—Ä—ã: {bad_images_list[:5]}")
print("=" * 40)

print(f"–ê–ù–ê–õ–ò–ó –û–ü–ê–°–ù–´–• –û–ü–ò–°–ê–ù–ò–ô:")
print("=" * 40)

if len(dangerous_descriptions) > 0:
    print("–¢–æ–ø-10 –æ–ø–∞—Å–Ω—ã—Ö –æ–ø–∏—Å–∞–Ω–∏–π:")
    for i, row in dangerous_descriptions.head(10).iterrows():
        print(f"  {i+1}. '{row['description_text'][:60]}...'")
        print(f"     –ò–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ: {row['image_file']}")
        print(f"     Approval: {row['approval_rate']:.2f}, Expert: {row['expert_mean']:.1f}")
        print(f"     –ü—Ä–∏—á–∏–Ω–∞: {row['reason']}")
        print()
````

````python
expert_available = master_df_cleaned[['expert1', 'expert2', 'expert3']].notna().any(axis=1).sum()
crowd_available = master_df_cleaned[['approval_rate', 'confirmed_count', 'rejected_count']].notna().any(axis=1).sum()
both_available = master_df_cleaned[
    master_df_cleaned[['expert1', 'expert2', 'expert3']].notna().any(axis=1) & 
    master_df_cleaned[['approval_rate', 'confirmed_count', 'rejected_count']].notna().any(axis=1)
].shape[0]

print(f"–° —ç–∫—Å–ø–µ—Ä—Ç–Ω—ã–º–∏ –æ—Ü–µ–Ω–∫–∞–º–∏: {expert_available} ({expert_available/len(master_df_cleaned)*100:.1f}%)")
print(f"–° –∫—Ä–∞—É–¥—Å–æ—Ä—Å–∏–Ω–≥–æ–≤—ã–º–∏ –æ—Ü–µ–Ω–∫–∞–º–∏: {crowd_available} ({crowd_available/len(master_df_cleaned)*100:.1f}%)")
print(f"–° –æ–±–æ–∏–º–∏ —Ç–∏–ø–∞–º–∏ –æ—Ü–µ–Ω–æ–∫: {both_available} ({both_available/len(master_df_cleaned)*100:.1f}%)")
````

````python
def aggregate_expert_scores_robust(df):
    aggregated_scores = []
    aggregation_methods = []  
    
    for idx, row in df.iterrows():
        expert_scores = []
        for expert_col in ['expert1', 'expert2', 'expert3']:
            if not pd.isna(row[expert_col]):
                expert_scores.append(row[expert_col])
        
        # –°–ª—É—á–∞–π 1: –ï—Å—Ç—å –≤—Å–µ 3 –æ—Ü–µ–Ω–∫–∏
        if len(expert_scores) == 3:
            unique_scores, counts = np.unique(expert_scores, return_counts=True)
            if np.max(counts) >= 2: 
                final_score = unique_scores[np.argmax(counts)]
                aggregated_scores.append(final_score)
                aggregation_methods.append('majority_3')
            else:  
                final_score = np.median(expert_scores)
                aggregated_scores.append(final_score)
                aggregation_methods.append('median_3')
        
        # –°–ª—É—á–∞–π 2: –ï—Å—Ç—å 2 –æ—Ü–µ–Ω–∫–∏
        elif len(expert_scores) == 2:
            if expert_scores[0] == expert_scores[1]: 
                final_score = expert_scores[0]
                aggregated_scores.append(final_score)
                aggregation_methods.append('agreement_2')
            else:  
                final_score = np.mean(expert_scores)
                aggregated_scores.append(final_score)
                aggregation_methods.append('average_2')
        
        # –°–ª—É—á–∞–π 3: –ï—Å—Ç—å 1 –æ—Ü–µ–Ω–∫–∞
        elif len(expert_scores) == 1:
            final_score = expert_scores[0]
            aggregated_scores.append(final_score)
            aggregation_methods.append('single_1')
        
        # –°–ª—É—á–∞–π 4: –ù–µ—Ç –æ—Ü–µ–Ω–æ–∫
        else:
            aggregated_scores.append(np.nan)
            aggregation_methods.append('none')
    
    return aggregated_scores, aggregation_methods
````

````python
expert_scores_agg, agg_methods = aggregate_expert_scores_robust(master_df_cleaned)
master_df_cleaned['expert_score_aggregated'] = expert_scores_agg
master_df_cleaned['aggregation_method'] = agg_methods

print("–†–µ–∑—É–ª—å—Ç–∞—Ç—ã –∞–≥—Ä–µ–≥–∞—Ü–∏–∏ —ç–∫—Å–ø–µ—Ä—Ç–Ω—ã—Ö –æ—Ü–µ–Ω–æ–∫:")
method_counts = pd.Series(agg_methods).value_counts()
for method, count in method_counts.items():
    print(f"  {method}: {count} –∑–∞–ø–∏—Å–µ–π ({count/len(master_df_cleaned)*100:.1f}%)")

print(f"–í—Å–µ–≥–æ –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–æ: {master_df_cleaned['expert_score_aggregated'].notna().sum()} –∑–∞–ø–∏—Å–µ–π")
````

**–í—ã–≤–æ–¥—ã –¥–ª—è –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏**
–ú—ã –∏–º–µ–µ–º –¥–µ–ª–æ —Å–æ —Å–ª–æ–∂–Ω–æ–π –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ—Å—Ç—å—é: –¥–æ–º–∏–Ω–∏—Ä–æ–≤–∞–Ω–∏–µ –Ω—É–ª–µ–π –∏ –Ω–∏–∑–∫–∏—Ö –æ—Ü–µ–Ω–æ–∫ –ø—Ä–∏ –Ω–∞–ª–∏—á–∏–∏ –Ω–µ–±–æ–ª—å—à–æ–≥–æ, –Ω–æ –≤–∞–∂–Ω–æ–≥–æ "—Ö–≤–æ—Å—Ç–∞" –≤—ã—Å–æ–∫–∏—Ö –æ—Ü–µ–Ω–æ–∫. –ú–æ–¥–µ–ª—å, –æ–±—É—á–µ–Ω–Ω–∞—è –±–µ–∑ —É—á–µ—Ç–∞ —ç—Ç–æ–≥–æ, –±—É–¥–µ—Ç –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞—Ç—å 0 –¥–ª—è –ø–æ—á—Ç–∏ –ª—é–±–æ–≥–æ –ø—Ä–∏–º–µ—Ä–∞ –∏ –≤—Å–µ —Ä–∞–≤–Ω–æ –ø–æ–ª—É—á–∏—Ç –≤—ã—Å–æ–∫—É—é —Ç–æ—á–Ω–æ—Å—Ç—å, –Ω–æ –±—É–¥–µ—Ç –±–µ—Å–ø–æ–ª–µ–∑–Ω–∞ –Ω–∞ –ø—Ä–∞–∫—Ç–∏–∫–µ. 

# –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—É—é –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫–∏

## –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è

````python
X_raw = master_df_cleaned[['image_file', 'description_id', 'description_text']]
y = master_df_cleaned['final_target'].values
groups = master_df_cleaned['image_file'].values

print(f"–í—Å–µ–≥–æ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è: {len(X_raw)}")
print(f"–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π: {len(np.unique(groups))}")
````

## –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ —Å —É—á–µ—Ç–æ–º –≥—Ä—É–ø–ø (–æ–¥–Ω–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ —Ç–æ–ª—å–∫–æ –≤ –æ–¥–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ)

````python
gss = GroupShuffleSplit(n_splits=1, train_size=0.7, random_state=RANDOM_STATE)
train_indices, valid_indices = next(gss.split(X=X_raw, y=y, groups=groups))

train_df = master_df_cleaned.iloc[train_indices].copy()
valid_df = master_df_cleaned.iloc[valid_indices].copy()

print(f"\n–†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è:")
print(f"–¢—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–∞—è –≤—ã–±–æ—Ä–∫–∞: {len(train_df)} –∑–∞–ø–∏—Å–µ–π")
print(f"–í–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–∞—è –≤—ã–±–æ—Ä–∫–∞: {len(valid_df)} –∑–∞–ø–∏—Å–µ–π")
print(f"–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –≤ train: {train_df['image_file'].nunique()}")
print(f"–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –≤ valid: {valid_df['image_file'].nunique()}")
````

## –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–π –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –º–µ–∂–¥—É train –∏ valid

````python
train_images_set = set(train_df['image_file'].unique())
valid_images_set = set(valid_df['image_file'].unique())
intersection = train_images_set & valid_images_set
print(f"–ü–µ—Ä–µ—Å–µ–∫–∞—é—â–∏—Ö—Å—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π: {len(intersection)}")
````

- Train: 3730 –∑–∞–ø–∏—Å–µ–π / 640 –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π = 5.83 –æ–ø–∏—Å–∞–Ω–∏—è –Ω–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ
- Valid: 1600 –∑–∞–ø–∏—Å–µ–π / 275 –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π = 5.82 –æ–ø–∏—Å–∞–Ω–∏—è –Ω–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ

–°—Ç—Ä—É–∫—Ç—É—Ä–∞ —Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω–∞—è: –≤ –æ–±–µ–∏—Ö –≤—ã–±–æ—Ä–∫–∞—Ö –ø—Ä–∏–º–µ—Ä–Ω–æ –ø–æ 5-6 –æ–ø–∏—Å–∞–Ω–∏–π –Ω–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ.

**–ü–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω—ã–µ –≤—ã–≥–æ–¥—ã –¥–ª—è –æ–ø—É—á–µ–Ω–∏—è:**
- **Train set** –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –±–æ–ª—å—à–æ–π (3,730 –ø—Ä–∏–º–µ—Ä–æ–≤) –¥–ª—è –æ–±—É—á–µ–Ω–∏—è –∫–∞—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π

- **Valid set** —Ä–µ–ø—Ä–µ–∑–µ–Ω—Ç–∞—Ç–∏–≤–µ–Ω (1,600 –ø—Ä–∏–º–µ—Ä–æ–≤) –¥–ª—è –Ω–∞–¥–µ–∂–Ω–æ–π –æ—Ü–µ–Ω–∫–∏

- –ì—Ä—É–ø–ø–æ–≤–æ–µ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–µ–¥–æ—Ç–≤—Ä–∞—â–∞–µ—Ç overfitting –Ω–∞ –∫–æ–Ω–∫—Ä–µ—Ç–Ω—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è—Ö

# –í–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è —Ç–µ–∫—Å—Ç–æ–≤ –∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π

## –í–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è —Ç–µ–∫—Å—Ç–æ–≤ —Å Sentence-Transformers

````python
print("–ó–∞–≥—Ä—É–∑–∫–∞ –º–æ–¥–µ–ª–∏ all-MiniLM-L6-v2...")
text_model = SentenceTransformer('all-MiniLM-L6-v2')

valid_texts = ["This is a test sentence.", "Another example text."]
valid_embeddings = text_model.encode(valid_texts)
print(f"–í–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–∞—è –≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è: {valid_embeddings.shape}")

print("–ú–æ–¥–µ–ª—å —É—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∂–µ–Ω–∞ –∏ –≥–æ—Ç–æ–≤–∞ –∫ —Ä–∞–±–æ—Ç–µ")
````

````python
train_texts = train_df['description_text'].tolist()
print(f"–í–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è {len(train_texts)} —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö —Ç–µ–∫—Å—Ç–æ–≤...")

train_text_embeddings = text_model.encode(
    train_texts,
    show_progress_bar=True,
    batch_size=32,
    convert_to_tensor=False,
    normalize_embeddings=True
)

print(f"–¢—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã–µ —Ç–µ–∫—Å—Ç–æ–≤—ã–µ —ç–º–±–µ–¥–¥–∏–Ω–≥–∏: {train_text_embeddings.shape}")
print(f"–î–∏–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω–∏–π: [{train_text_embeddings.min():.3f}, {train_text_embeddings.max():.3f}]")
````

## –í–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π —Å ResNet-18

````python
def get_image_embedding(image_path, model, transform):
    try:
        img = Image.open(image_path).convert('RGB')
        img_tensor = transform(img).unsqueeze(0).to(device)
        
        with torch.no_grad():
            embedding = model(img_tensor)
        
        return embedding.squeeze().cpu().numpy()
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –æ–±—Ä–∞–±–æ—Ç–∫–µ {image_path}: {e}")
        return None

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"–ò—Å–ø–æ–ª—å–∑—É–µ–º–æ–µ —É—Å—Ç—Ä–æ–π—Å—Ç–≤–æ: {device}")

resnet = models.resnet18(pretrained=True)
resnet = torch.nn.Sequential(*(list(resnet.children())[:-1]))
resnet = resnet.to(device)
resnet.eval()

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

print("–ú–æ–¥–µ–ª—å ResNet-18 –∑–∞–≥—Ä—É–∂–µ–Ω–∞ –∏ –≥–æ—Ç–æ–≤–∞ –∫ —Ä–∞–±–æ—Ç–µ")
````

````python
print("\n–í–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π")
print("=" * 60)

def vectorize_images_batch(image_files, images_dir, model, transform, batch_name=""):
    embeddings_dict = {}
    missing_images = []
    
    print(f"–í–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è {len(image_files)} –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π {batch_name}...")
    
    for i, img_file in enumerate(image_files):
        img_path = os.path.join(images_dir, img_file)
        embedding = get_image_embedding(img_path, model, transform)
        
        if embedding is not None:
            embeddings_dict[img_file] = embedding
        else:
            missing_images.append(img_file)
        
        if (i + 1) % 50 == 0:
            print(f"  –û–±—Ä–∞–±–æ—Ç–∞–Ω–æ {i + 1}/{len(image_files)}")
    
    if missing_images:
        print(f"–ü—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏–µ: {len(missing_images)} –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –Ω–µ –Ω–∞–π–¥–µ–Ω–æ")
    else:
        print(f"–í—Å–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è —É—Å–ø–µ—à–Ω–æ –æ–±—Ä–∞–±–æ—Ç–∞–Ω—ã")
    
    return embeddings_dict

train_images = train_df['image_file'].unique()
train_image_embeddings = vectorize_images_batch(
    train_images, TRAIN_IMAGES_DIR, resnet, transform, "train"
)

print(f"–¢—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è: {len(train_image_embeddings)}/{len(train_images)}")

with open('train_image_embeddings.pkl', 'wb') as f:
    pickle.dump(train_image_embeddings, f)
print("‚úì –¢—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã–µ —ç–º–±–µ–¥–¥–∏–Ω–≥–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ train_image_embeddings.pkl")
````

````python
def create_normalized_multimodal_features(text_embeddings_norm, image_embeddings_norm):
    multimodal_features = []
    
    for i in range(len(text_embeddings_norm)):
        combined_emb = np.concatenate([text_embeddings_norm[i], image_embeddings_norm[i]])
        multimodal_features.append(combined_emb)
    
    return np.array(multimodal_features)
````

- –ß—É–≤—Å—Ç–≤–∏—Ç–µ–ª–µ–Ω –∫ –≤—ã–±–æ—Ä—É –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –∏ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—é

- –°–æ—Ö—Ä–∞–Ω—è–µ—Ç —Ä–∞–∑–Ω–æ–æ–±—Ä–∞–∑–∏–µ –ø–æ–∏—Å–∫–∞ –ø—Ä–∏ —Å–æ–∫—Ä–∞—â–µ–Ω–∏–∏ –≤—Ä–µ–º–µ–Ω–∏

**–ü—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–∞ –ø–æ–¥—Ö–æ–¥–∞:**

- –ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∞—è –æ—Å—É—â–µ—Å—Ç–≤–∏–º–æ—Å—Ç—å –æ–±—É—á–µ–Ω–∏—è

- –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏ –Ω–∞–π—Ç–∏ –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã

- –£—á–µ—Ç –≤—ã—á–∏—Å–ª–∏—Ç–µ–ª—å–Ω–æ–π —Å–ª–æ–∂–Ω–æ—Å—Ç–∏ SVR –Ω–∞ –Ω–∞—à–∏—Ö –¥–∞–Ω–Ω—ã—Ö

***–î–ª—è –æ—Ü–µ–Ω–∫–∏ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–µ–π –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è –Ω–∞–±–æ—Ä–æ–º –º–µ—Ç—Ä–∏–∫:***

1. –î–ª—è –æ—Ü–µ–Ω–∫–∏ —Ç–æ—á–Ω–æ—Å—Ç–∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ–π —Ä–µ–ª–µ–≤–∞–Ω—Ç–Ω–æ—Å—Ç–∏ –∏—Å–ø–æ–ª—å–∑—É–µ–º **MAE**.
    
    - –ú–µ—Ç—Ä–∏–∫–∞ –ª–µ–≥–∫–æ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä—É–µ—Ç—Å—è.
    - –£—Å—Ç–æ–π—á–∏–≤–∞ –∫ –≤—ã–±—Ä–æ—Å–∞–º, –∞ –≤ –Ω–∞—à–∏—Ö –¥–∞–Ω–Ω—ã—Ö –µ—Å—Ç—å –∫—Ä–∞–π–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è, —Ç.–µ. –≤—ã–±—Ä–æ—Å—ã –≤–µ—Ä–æ—è—Ç–Ω—ã.

2. –î–ª—è –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ —Ä–∞–Ω–∂–∏—Ä–æ–≤–∞–Ω–∏—è –¥–ª—è –±–∏–∑–Ω–µ—Å-–∑–∞–¥–∞—á–∏ –∏—Å–ø–æ–ª—å–∑—É–µ–º **Precision@K**.

3. –î–ª—è –æ—Ü–µ–Ω–∫–∏ –æ–±—ä—è—Å–Ω—è—é—â–µ–π —Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–∏ –∏—Å–ø–æ–ª—å–∑—É–µ–º **R¬≤** - –æ–Ω–∞ –ø–æ–∫–∞–∂–µ—Ç, –∫–∞–∫–∞—è –¥–æ–ª—è –¥–∏—Å–ø–µ—Ä—Å–∏–∏ —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –æ–±—ä—è—Å–Ω—è–µ—Ç—Å—è –º–æ–¥–µ–ª—å—é.

## –ò—Å–ø–æ–ª—å–∑—É–µ–º –ø–æ–¥–≥–æ—Ç–æ–≤–ª–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

````python
print(f"–†–∞–∑–º–µ—Ä—ã –¥–∞–Ω–Ω—ã—Ö:")
print(f"X_train_scaled: {X_train_scaled.shape}")
print(f"X_valid_scaled: {X_valid_scaled.shape}")
print(f"y_train: {y_train.shape}")
print(f"y_valid: {y_valid.shape}")
````

## –û–ø—Ä–µ–¥–µ–ª—è–µ–º –º–æ–¥–µ–ª–∏ –∏ –ø–æ–¥—Ö–æ–¥ –∫ –æ–±—É—á–µ–Ω–∏—é

````python
models_advanced = {
    'LinearRegression': LinearRegression(),
    'LightGBM': LGBMRegressor(random_state=RANDOM_STATE, verbosity=-1),
    'SVR': SVR()
}

param_distributions = {
    'SVR': {
        'C': stats.uniform(0.1, 10),  
        'epsilon': stats.uniform(0.01, 0.2),  
        'kernel': ['rbf', 'linear']
    }
}

param_grids = {
    'LinearRegression': {},
    'LightGBM': {
        'n_estimators': [100, 150, 200],
        'learning_rate': [0.05, 0.1, 0.15],
        'num_leaves': [31, 63, 127],
        'reg_alpha': [0, 0.1, 0.5],
        'reg_lambda': [0, 0.1, 0.5]
    },
    'SVR': {
        'C': stats.uniform(0.1, 10),  
        'epsilon': stats.uniform(0.01, 0.2),
        'kernel': ['rbf', 'linear']
    }
}
````

````python
def precision_at_k(y_true, y_pred, k=100, threshold_quantile=0.8):
    threshold = np.quantile(y_true, threshold_quantile)
    high_relevance_true = y_true >= threshold
    top_k_indices = np.argsort(y_pred)[-k:]
    precision = np.sum(high_relevance_true[top_k_indices]) / k
    return precision

def evaluate_model(model, X, y, model_name=""):
    y_pred = model.predict(X)
    metrics = {
        'MAE': mean_absolute_error(y, y_pred),
        'R2': r2_score(y, y_pred),
        'Precision@100': precision_at_k(y, y_pred, k=100),
        'Precision@50': precision_at_k(y, y_pred, k=50)
    }
    
    if model_name:
        print(f"\n{model_name} –†–µ–∑—É–ª—å—Ç–∞—Ç—ã:")
        print("-" * 40)
    
    for metric, value in metrics.items():
        print(f"{metric}: {value:.4f}")
    
    return metrics, y_pred
````

## –û–±—É—á–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–∏—Ö –º–æ–¥–µ–ª–µ–π —Å GroupShuffleSplit

````python
train_groups = train_df['image_file'].values
print(f"–†–∞–∑–º–µ—Ä train_groups: {len(train_groups)}")
print(f"–†–∞–∑–º–µ—Ä X_train_scaled: {X_train_scaled.shape[0]}")
print(f"–†–∞–∑–º–µ—Ä y_train: {len(y_train)}")

assert len(train_groups) == X_train_scaled.shape[0] == len(y_train), "–†–∞–∑–º–µ—Ä—ã –¥–∞–Ω–Ω—ã—Ö –Ω–µ —Å–æ–≤–ø–∞–¥–∞—é—Ç!"

warnings.filterwarnings("ignore", category=UserWarning, module="sklearn")

group_kfold = GroupKFold(n_splits=CV)
best_models = {}
results = {}

for model_name, model in models_advanced.items():
    print(f"\n{'='*60}")
    print(f"–û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏: {model_name}")
    print(f"{'='*60}")
    
    if model_name == 'LinearRegression':
        print("–û–±—É—á–µ–Ω–∏–µ LinearRegression...")
        %time model.fit(X_train_scaled, y_train)
        best_models[model_name] = model
        
    elif model_name == 'SVR':
        random_search = RandomizedSearchCV(
            estimator=model,
            param_distributions=param_distributions[model_name],
            n_iter=8,
            scoring='neg_mean_absolute_error',
            cv=group_kfold, 
            n_jobs=-1,
            random_state=RANDOM_STATE,
            verbose=1
        )
        
        print("–°–ª—É—á–∞–π–Ω—ã–π –ø–æ–∏—Å–∫ –ª—É—á—à–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è SVR (—Å GroupKFold)...")
        %time random_search.fit(X_train_scaled, y_train, groups=train_groups) 
        best_models[model_name] = random_search.best_estimator_
        
    else:
        grid_search = GridSearchCV(
            estimator=model,
            param_grid=param_grids[model_name],
            scoring='neg_mean_absolute_error',
            cv=group_kfold,  
            n_jobs=-1,
            verbose=0
        )
        
        print("–ü–æ–∏—Å–∫ –ª—É—á—à–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ (—Å GroupKFold)...")
        %time grid_search.fit(X_train_scaled, y_train, groups=train_groups)  
        best_models[model_name] = grid_search.best_estimator_
    
    with warnings.catch_warnings():
        warnings.filterwarnings("ignore", category=UserWarning, module="sklearn")
        metrics, y_pred = evaluate_model(best_models[model_name], X_valid_scaled, y_valid, model_name)
    
    results[model_name] = metrics
    
    if model_name != 'LinearRegression':
        if model_name == 'SVR':
            print(f"–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã: {random_search.best_params_}")
        else:
            print(f"–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã: {grid_search.best_params_}")
    
    print(f"MAE: {metrics['MAE']:.4f}")
    print(f"R2: {metrics['R2']:.4f}")
    print(f"Precision@100: {metrics['Precision@100']:.4f}")
````

## –û—Ç–¥–µ–ª—å–Ω–æ–µ –æ–±—É—á–µ–Ω–∏–µ –Ω–µ–π—Ä–æ–Ω–Ω–æ–π —Å–µ—Ç–∏ —Å PyTorch

````python
class MultimodalNN(nn.Module):
    def __init__(self, input_dim, hidden_dims=[512, 256, 128], dropout=0.3):
        super(MultimodalNN, self).__init__()
        layers = []
        prev_dim = input_dim
        
        for hidden_dim in hidden_dims:
            layers.extend([
                nn.Linear(prev_dim, hidden_dim),
                nn.BatchNorm1d(hidden_dim),
                nn.ReLU(),
                nn.Dropout(dropout)
            ])
            prev_dim = hidden_dim
        
        layers.append(nn.Linear(prev_dim, 1))
        self.network = nn.Sequential(*layers)
    
    def forward(self, x):
        return self.network(x).squeeze()

print(f"\n{'='*60}")
print(f"–û–±—É—á–µ–Ω–∏–µ –Ω–µ–π—Ä–æ–Ω–Ω–æ–π —Å–µ—Ç–∏: MultimodalNN")
print(f"{'='*60}")

X_train_tensor = torch.FloatTensor(X_train_scaled)
y_train_tensor = torch.FloatTensor(y_train)
X_valid_tensor = torch.FloatTensor(X_valid_scaled)
y_valid_tensor = torch.FloatTensor(y_valid)

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)

model_nn = MultimodalNN(input_dim=X_train_scaled.shape[1])
criterion = nn.L1Loss()  # MAE loss
optimizer = optim.Adam(model_nn.parameters(), lr=0.001, weight_decay=1e-5)

model_nn.train()
print("–û–±—É—á–µ–Ω–∏–µ –Ω–µ–π—Ä–æ–Ω–Ω–æ–π —Å–µ—Ç–∏...")
start_time = time.time()

for epoch in range(100):
    epoch_loss = 0
    for batch_X, batch_y in train_loader:
        optimizer.zero_grad()
        outputs = model_nn(batch_X)
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    
    if (epoch + 1) % 20 == 0:
        print(f"–≠–ø–æ—Ö–∞ {epoch+1}, Loss: {epoch_loss/len(train_loader):.4f}")

training_time = time.time() - start_time
print(f"–í—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è –Ω–µ–π—Ä–æ–Ω–Ω–æ–π —Å–µ—Ç–∏: {training_time:.2f} —Å–µ–∫—É–Ω–¥")

model_nn.eval()
with torch.no_grad():
    y_pred_nn = model_nn(X_valid_tensor).numpy()

metrics_nn = {
    'MAE': mean_absolute_error(y_valid, y_pred_nn),
    'R2': r2_score(y_valid, y_pred_nn),
    'Precision@100': precision_at_k(y_valid, y_pred_nn, k=100)
}

results['MultimodalNN'] = metrics_nn
best_models['MultimodalNN'] = model_nn

print(f"MAE: {metrics_nn['MAE']:.4f}")
print(f"R2: {metrics_nn['R2']:.4f}")
print(f"Precision@100: {metrics_nn['Precision@100']:.4f}")
````

## –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

````python
comparison_df = pd.DataFrame(results).T
comparison_df = comparison_df.sort_values('MAE')

print(comparison_df)

best_model_name = comparison_df.index[0]
best_model = best_models[best_model_name]
best_metrics = results[best_model_name]

print(f"\n{'='*60}")
print(f"–õ–£–ß–®–ê–Ø –ú–û–î–ï–õ–¨: {best_model_name}")
print(f"{'='*60}")
print(f"MAE: {best_metrics['MAE']:.4f}")
print(f"R2: {best_metrics['R2']:.4f}")
print(f"Precision@100: {best_metrics['Precision@100']:.4f}")
````

SVR –ø—Ä–æ–¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä–æ–≤–∞–ª–∞ –∏–¥–µ–∞–ª—å–Ω—É—é —Ç–æ—á–Ω–æ—Å—Ç—å –∏ –ª—É—á—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç R¬≤ - –≤–æ–∑–º–æ–∂–Ω–æ, –æ–Ω–∞ –ª—É—á—à–µ —É–ª–∞–≤–ª–∏–≤–∞–µ—Ç –æ–±—â–∏–µ –∑–∞–∫–æ–Ω–æ–º–µ—Ä–Ω–æ—Å—Ç–∏. –¢–µ–º –Ω–µ –º–µ–Ω–µ–µ –º–æ–∂–Ω–æ —Å—á–∏—Ç–∞—Ç—å, —á—Ç–æ –æ–±—Ä–∞—â–µ–Ω–∏–µ –∫ —ç—Ç–æ–π –º–æ–¥–µ–ª–∏ —Å–µ–±—è –Ω–µ –æ–ø—Ä–∞–≤–¥–∞–ª–æ. –ï–µ –æ–±—É—á–µ–Ω–∏–µ –∑–∞–Ω—è–ª–æ —á–∞—Å —Å –ª–∏—à–Ω–∏–º.

MultimodalNN –æ–∫–∞–∑–∞–ª–∞—Å—å —á—É—Ç—å –º–µ–Ω–µ–µ —Ç–æ—á–Ω–æ–π, –∑–∞—Ç–æ MAE –Ω–∞–∏–º–µ–Ω—å—à—É—é –æ—à–∏–±–∫—É, –Ω–∞ 10% –ª—É—á—à–µ –±–ª–∏–∂–∞–π—à–µ–≥–æ –∫–æ–Ω–∫—É—Ä–µ–Ω—Ç–∞. –ú–æ–∂–Ω–æ —Å–∫–∞–∑–∞—Ç—å, —á—Ç–æ MultimodalNN –±–∞–ª–∞–Ω—Å–∏—Ä—É–µ—Ç –º–µ–∂–¥—É —Ç–æ—á–Ω–æ—Å—Ç—å—é –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –∏ –∫–∞—á–µ—Å—Ç–≤–æ–º —Ä–∞–Ω–∂–∏—Ä–æ–≤–∞–Ω–∏—è. –ù–æ –µ–µ –æ–±—É—á–µ–Ω–∏–µ –∑–∞–Ω—è–ª–æ —á—É—Ç—å –º–µ–Ω—å—à–µ 20 —Å–µ–∫—É–Ω–¥.

````python
if best_model_name == 'MultimodalNN':
    torch.save({
        'model_state_dict': best_model.state_dict(),
        'input_dim': X_train_scaled.shape[1]
    }, f'best_{best_model_name}.pth')
else:
    joblib.dump(best_model, f'best_{best_model_name}.pkl')

print(f"–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∞ –∫–∞–∫: best_{best_model_name}.pth" if best_model_name == 'MultimodalNN' else f"best_{best_model_name}.pkl")
````

group_kfold = GroupKFold(n_splits=CV)

### –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ —ç–º–±–µ–¥–¥–∏–Ω–≥–æ–≤ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –¥–ª—è –ø–æ–∏—Å–∫–∞

````python
def vectorize_test_images(image_files, model, transform, batch_name=""):
    embeddings_dict = {}
    missing_images = []
    
    print(f"–í–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è {len(image_files)} –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π {batch_name}...")
    print(f"–ü—É—Ç—å –∫ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º: {TEST_IMAGES_DIR}")
    
    for i, img_file in enumerate(image_files):
        img_path = os.path.join(TEST_IMAGES_DIR, img_file)
        
        if not os.path.exists(img_path):
            print(f"–§–∞–π–ª –Ω–µ –Ω–∞–π–¥–µ–Ω: {img_path}")
            missing_images.append(img_file)
            continue
            
        embedding = get_image_embedding(img_path, model, transform)
        
        if embedding is not None:
            embeddings_dict[img_file] = embedding
        else:
            missing_images.append(img_file)
        
        if (i + 1) % 20 == 0:
            print(f"  –û–±—Ä–∞–±–æ—Ç–∞–Ω–æ {i + 1}/{len(image_files)}")
    
    if missing_images:
        print(f"–ü—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏–µ: {len(missing_images)} –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –Ω–µ –Ω–∞–π–¥–µ–Ω–æ –∏–ª–∏ –Ω–µ –æ–±—Ä–∞–±–æ—Ç–∞–Ω–æ")
        print(f"–ü–µ—Ä–≤—ã–µ 5 –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö: {missing_images[:5]}")
    else:
        print(f"–í—Å–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è —É—Å–ø–µ—à–Ω–æ –æ–±—Ä–∞–±–æ—Ç–∞–Ω—ã")
    
    return embeddings_dict

def get_image_embedding(image_path, model, transform):
    try:
        img = Image.open(image_path).convert('RGB')
        img_tensor = transform(img).unsqueeze(0).to(device)
        
        with torch.no_grad():
            embedding = model(img_tensor)
            embedding = F.adaptive_avg_pool2d(embedding, (1, 1))
            embedding = embedding.squeeze(-1).squeeze(-1)
        
        return embedding.cpu().numpy()
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –æ–±—Ä–∞–±–æ—Ç–∫–µ {image_path}: {e}")
        return None

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"–ò—Å–ø–æ–ª—å–∑—É–µ–º–æ–µ —É—Å—Ç—Ä–æ–π—Å—Ç–≤–æ: {device}")
````

### –°–æ–∑–¥–∞–Ω–∏–µ –ø–æ–∏—Å–∫–æ–≤–æ–π —Å–∏—Å—Ç–µ–º—ã

````python
def search_system(query_text, top_k=5):
    children_keywords = [
        'child', 'children', 'kid', 'kids', 'boy', 'boys', 'girl', 'girls',
        'teenager', 'teen', 'baby', 'babies', 'toddler', 'young boy', 'young girl',
        'little boy', 'little girl', 'schoolboy', 'schoolgirl', 'youngster',
        'minor', 'youth', 'adolescent', 'juvenile', 'baby boy', 'baby girl'
    ]
    
    query_lower = query_text.lower()
    has_dangerous_content = any(keyword in query_lower for keyword in children_keywords)
    
    if has_dangerous_content:
        return {
            'status': 'blocked',
            'message': 'This image is unavailable in your country in compliance with local laws',
            'query': query_text
        }
    
    try:
        query_embedding = text_model.encode([query_text], convert_to_tensor=False, normalize_embeddings=True)[0]
        query_embedding_normalized = text_normalizer.transform([query_embedding])[0]
        
        multimodal_features = []
        for img_embedding in image_embeddings_normalized:
            combined = np.concatenate([query_embedding_normalized, img_embedding])
            multimodal_features.append(combined)
        
        multimodal_array = np.array(multimodal_features)
        
        if best_model_name == 'MultimodalNN':
            with torch.no_grad():
                multimodal_tensor = torch.FloatTensor(multimodal_array)
                relevance_scores = best_model(multimodal_tensor).numpy()
        else:
            relevance_scores = best_model.predict(multimodal_array)
        
        sorted_indices = np.argsort(relevance_scores)[::-1][:top_k]
        
        results = []
        for idx in sorted_indices:
            results.append({
                'image_name': image_names[idx],
                'relevance_score': float(relevance_scores[idx]),
                'rank': len(results) + 1
            })
        
        return {
            'status': 'success',
            'query': query_text,
            'top_k': top_k,
            'results': results
        }
        
    except Exception as e:
        return {
            'status': 'error',
            'message': f'Search error: {str(e)}',
            'query': query_text
        }

test_queries_sample = test_queries.sample(5, random_state=42)

for idx, row in test_queries_sample.iterrows():
    print(f"\n{'‚îÄ'*60}")
    print(f"–¢–µ—Å—Ç–æ–≤—ã–π –∑–∞–ø—Ä–æ—Å: '{row['query_text']}'")
    print(f"–û–∂–∏–¥–∞–µ–º–æ–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ: {row['image']}")
    
    result = search_system(row['query_text'], top_k=3)
    
    if result['status'] == 'blocked':
        print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {result['message']}")
    elif result['status'] == 'success':
        print("–ù–∞–π–¥–µ–Ω–Ω—ã–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è:")
        for res in result['results']:
            match_indicator = " ‚úì" if res['image_name'] == row['image'] else ""
            print(f"  {res['rank']}. {res['image_name']} (score: {res['relevance_score']:.3f}){match_indicator}")
    else:
        print(f"–û—à–∏–±–∫–∞: {result['message']}")
````

````python
if best_model_name == 'MultimodalNN':
    model_nn.eval()
    best_model = model_nn
    print("–ó–∞–≥—Ä—É–∂–µ–Ω–∞ –Ω–µ–π—Ä–æ–Ω–Ω–∞—è —Å–µ—Ç—å MultimodalNN")
else:
    best_model = best_models[best_model_name]
    print(f"–ó–∞–≥—Ä—É–∂–µ–Ω–∞ –º–æ–¥–µ–ª—å: {best_model_name}")

def search_system(query_text, top_k=5):
    children_keywords = [
        'child', 'children', 'kid', 'kids', 'boy', 'boys', 'girl', 'girls',
        'teenager', 'teen', 'baby', 'babies', 'toddler', 'young boy', 'young girl',
        'little boy', 'little girl', 'schoolboy', 'schoolgirl', 'youngster',
        'minor', 'youth', 'adolescent', 'juvenile', 'baby boy', 'baby girl'
    ]
    
    query_lower = query_text.lower()
    has_dangerous_content = any(keyword in query_lower for keyword in children_keywords)
    
    if has_dangerous_content:
        return {
            'status': 'blocked',
            'message': 'This image is unavailable in your country in compliance with local laws',
            'query': query_text
        }
    
    try:
        query_embedding = text_model.encode([query_text], convert_to_tensor=False, normalize_embeddings=True)[0]
        query_embedding_normalized = text_normalizer.transform([query_embedding])[0]
        
        multimodal_features = []
        for img_embedding in image_embeddings_normalized:
            combined = np.concatenate([query_embedding_normalized, img_embedding])
            multimodal_features.append(combined)
        
        multimodal_array = np.array(multimodal_features)
        
        if best_model_name == 'MultimodalNN':
            with torch.no_grad():
                multimodal_tensor = torch.FloatTensor(multimodal_array)
                relevance_scores = best_model(multimodal_tensor).numpy()
        else:
            relevance_scores = best_model.predict(multimodal_array)
        
        sorted_indices = np.argsort(relevance_scores)[::-1][:top_k]
        
        results = []
        for idx in sorted_indices:
            results.append({
                'image_name': image_names[idx],
                'relevance_score': float(relevance_scores[idx]),
                'rank': len(results) + 1
            })
        
        return {
            'status': 'success',
            'query': query_text,
            'top_k': top_k,
            'results': results
        }
        
    except Exception as e:
        return {
            'status': 'error',
            'message': f'Search error: {str(e)}',
            'query': query_text
        }
````

### –ü–æ–∏—Å–∫ —Ä–µ–ª–µ–≤–∞–Ω—Ç–Ω—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –ø–æ —Ç–µ–∫—Å—Ç–æ–≤–æ–º—É –∑–∞–ø—Ä–æ—Å—É

````python
print("\n–í—ã–±–∏—Ä–∞–µ–º 10 —Å–ª—É—á–∞–π–Ω—ã—Ö –∑–∞–ø—Ä–æ—Å–æ–≤ –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è")
print("-" * 60)

test_sample = test_queries.sample(10, random_state=42)

search_results = []
correct_predictions = 0

print("–†–µ–∑—É–ª—å—Ç–∞—Ç—ã –ø–æ–∏—Å–∫–∞:")
print("-" * 80)

def display_search_result_image(image_filename, title, is_expected=False):
    try:
        image_path = os.path.join(TEST_IMAGES_DIR, image_filename)
        if os.path.exists(image_path):
            img = Image.open(image_path)
            
            plt.figure(figsize=(8, 6))
            plt.imshow(img)
            
            if is_expected:
                plt.title(f"–û–ñ–ò–î–ê–ï–ú–û–ï: {title}", fontsize=12, fontweight='bold', color='green')
                for spine in plt.gca().spines.values():
                    spine.set_edgecolor('green')
                    spine.set_linewidth(3)
            else:
                plt.title(f"–ù–ê–ô–î–ï–ù–û: {title}", fontsize=12)
                
            plt.axis('off')
            plt.tight_layout()
            plt.show()
            return True
        else:
            print(f"–ò–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –Ω–µ –Ω–∞–π–¥–µ–Ω–æ: {image_filename}")
            return False
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –∑–∞–≥—Ä—É–∑–∫–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è {image_filename}: {e}")
        return False

for idx, row in test_sample.iterrows():
    query_text = row['query_text']
    expected_image = row['image']
    
    result = search_system(query_text, top_k=5)
    
    if result['status'] == 'blocked':
        print(f"–ó–ê–ü–†–û–°: '{query_text}'")
        print(f"–†–ï–ó–£–õ–¨–¢–ê–¢: –ó–ê–ë–õ–û–ö–ò–†–û–í–ê–ù - {result['message']}")
        print(f"–û–ñ–ò–î–ê–ï–ú–û–ï –ò–ó–û–ë–†–ê–ñ–ï–ù–ò–ï: {expected_image}")
        search_results.append({
            'query': query_text,
            'expected_image': expected_image,
            'status': 'blocked',
            'found_image': None,
            'relevance_score': None
        })
        
    elif result['status'] == 'success':
        found_image = result['results'][0]['image_name']
        relevance_score = result['results'][0]['relevance_score']
        is_correct = (found_image == expected_image)
        
        if is_correct:
            correct_predictions += 1
        
        print(f"–ó–ê–ü–†–û–°: '{query_text}'")
        print(f"–ù–ê–ô–î–ï–ù–û: {found_image} (score: {relevance_score:.3f})")
        print(f"–û–ñ–ò–î–ê–ï–ú–û–ï: {expected_image}")
        print(f"–°–¢–ê–¢–£–°: {'–°–û–í–ü–ê–î–ï–ù–ò–ï' if is_correct else '–ù–ï–°–û–í–ü–ê–î–ï–ù–ò–ï'}")
        
        print("\n–ù–ê–ô–î–ï–ù–ù–û–ï –ò–ó–û–ë–†–ê–ñ–ï–ù–ò–ï:")
        display_search_result_image(found_image, f"Score: {relevance_score:.3f}")
        
        if not is_correct:
            print("\n–û–ñ–ò–î–ê–ï–ú–û–ï –ò–ó–û–ë–†–ê–ñ–ï–ù–ò–ï:")
            display_search_result_image(expected_image, "Expected", is_expected=True)
        
        print("–¢–æ–ø-3 —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã:")
        for i, res in enumerate(result['results'][:3]):
            match_indicator = " <-- –û–ñ–ò–î–ê–ï–ú–´–ô" if res['image_name'] == expected_image else ""
            print(f"  {i+1}. {res['image_name']} (score: {res['relevance_score']:.3f}){match_indicator}")
            
            if i > 0 and res['image_name'] != found_image and res['image_name'] != expected_image:
                display_search_result_image(res['image_name'], f"–¢–æ–ø-{i+1}: Score: {res['relevance_score']:.3f}")
        
        search_results.append({
            'query': query_text,
            'expected_image': expected_image,
            'status': 'success',
            'found_image': found_image,
            'relevance_score': relevance_score,
            'is_correct': is_correct
        })
    
    else:
        print(f"–ó–ê–ü–†–û–°: '{query_text}'")
        print(f"–û–®–ò–ë–ö–ê: {result['message']}")
        search_results.append({
            'query': query_text,
            'expected_image': expected_image,
            'status': 'error',
            'found_image': None,
            'relevance_score': None
        })
````
